```python
import pandas as pd
path="../data_jitter/"
```

```python
from scipy import interpolate
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import font_manager
import matplotlib.gridspec as gridspec
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.ticker import ScalarFormatter
mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
mss=120
fontssize=50
mpl.rcParams.update({
    'font.family': 'serif',
    'font.serif': ['Liberation Sans'],  # Add or remove font names as needed
    #'font.sans-serif':['Helvetica'],
    'font.size': fontssize,  # Adjust as needed
    #'font.weight': 'bold',
    "svg.fonttype": 'none',
    'text.usetex': False,
    'axes.linewidth' : 2,
    #'axes.labelweight': 'bold',
    'text.latex.preamble': r'\usepackage{amsmath} \usepackage{amsfonts} \usepackage{cmbright}',
    'xtick.labelsize' : fontssize, # fontsize of the x tick labels
    'ytick.labelsize' : fontssize, # fontsize of the y tick labels
})
ticks_font = font_manager.FontProperties(family='Liberation Sans', style='normal',
    size=fontssize, weight='bold', stretch='normal')
```

```python

quant1='msr'
quant2='mean_msrDSP'
```

```python

#jitter=1000 ## jitter of 50ms
for jitter in [5,10,50,100,500,1000]: #5 10 50 100 500 1000 milisecs
    data=pd.read_json(path+"Rat_21012_Rec_18_jitter_"+str(jitter)+"ms_data.json",orient="records")

    fig, ax1 = plt.subplots(figsize=(10,10))
    figname='../figures/jitterMSR_'+str(jitter)+'ms.pdf'


    ax1.scatter(data.query('loc=="SUB"')[quant1],data.query('loc=="SUB"')[quant2],
            label='SUB Neurons', marker='o', facecolors='none',edgecolors='b',s=mss)
    ax1.scatter(data.query('loc=="CA1"')[quant1],data.query('loc=="CA1"')[quant2],
            label='CA1 Neurons', marker='D', facecolors='none',edgecolors='r',s=mss)

    ax1.set_yscale('log')
    ax1.set_ylim(0.00001,5e-3)
    # ax1.text(-0.3, 1.0, r'(x$10^{-3}$)',
    #          transform=ax1.transAxes, family='Liberation Sans', style='normal',
    #     size=fontssize, weight='bold', stretch='normal',verticalalignment='top', bbox=dict(facecolor='white', alpha=0.5, linewidth=0))

    ax1.set_xlabel(quant1.upper()+"(bits$^2$)");
    ax1.set_ylabel("Mean Displacement");

    for tick in ax1.get_xticklabels():
        tick.set_fontproperties(ticks_font)
    for tick in ax1.get_yticklabels():
        tick.set_fontproperties(ticks_font)
        
    fig.savefig(figname,bbox_inches='tight',dpi=300)
    fig.clf()
```
